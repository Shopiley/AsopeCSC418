```python
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np

pic_map = {"anjola":"anjola.jpg", "fatima":"fatima.jpg", "francis":"francis.jpg", "keisha":"keisha.jpg", "lotanna":"lotanna.jpg", "oyinda":"oyinda.jpg", "pese":"pese.jpg", "sophia":"sophia.jpg", "wendy":"wendy.jpg", "zion":"zion.jpg" }

def addition_of_images(img1):
    img1 = cv2.imread('img/'+ img1)
    img2 = cv2.imread('img/sope.jpg')

    img1 = cv2.resize(img1, (500, 400))
    img2 = cv2.resize(img2, (500, 400))

    add_image = cv2.addWeighted(img1, 0.5, img2, 0.6, 0)

    cv2.imshow('add_image', add_image)

    if cv2.waitKey(0) & 0xff == 27:
        cv2.destroyAllWindows()


def adjust_brightness_contrast(img_):
    img = cv2.imread('img/'+ img_)

    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(img) 

    brightness = 5
    contrast = 1.5
    img2 = cv2.addWeighted(img, contrast, np.zeros(img.shape, img.dtype), 0, brightness)   

    cv2.imwrite(f'img/contrast_{img_.split(".", 1)[0]}.jpg', img2)

    plt.subplot(1, 2, 2)
    plt.title("Brightness & Contrast")

    plt.imshow(img2)
    plt.show()


def sharpening(img_):
    img = cv2.imread('img/'+ img_)
    
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(img)

    kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])

    sharpened_img = cv2.filter2D(img, -1, kernel)

    cv2.imwrite(f'img/sharpened_{img_.split(".", 1)[0]}.jpg', sharpened_img) 

    plt.subplot(1, 2, 2)
    plt.title("Sharpened")
    plt.imshow(sharpened_img)
    plt.show()


def removing_noise(img_):
    img = cv2.imread('img/'+ img_)

    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(img)

    filter_image = cv2.medianBlur(img, 15)

    cv2.imwrite(f'img/median_{img_.split(".", 1)[0]}.jpg', filter_image)

    plt.subplot(1, 2, 2)
    plt.title("Median Blur")
    plt.imshow(filter_image)
    plt.show()


def image_scaling(img_):
    img = cv2.imread('img/'+ img_)

    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(img)

    scaled_image = cv2.resize(img, None, fx=2, fy=2)

    cv2.imwrite(f'img/scaled_{img_.split(".", 1)[0]}.jpg', scaled_image)

    plt.subplot(1, 2, 2)
    plt.title("Scaled")
    plt.imshow(scaled_image)
    plt.show()


def inverse_transformation(img_):
    img = cv2.imread('img/'+ img_)

    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(img)

    # inversed_image = cv2.bitwise_not(img)
    inversed_image = 255 - img

    cv2.imwrite(f'img/inversed_{img_.split(".", 1)[0]}.jpg', inversed_image)

    plt.subplot(1, 2, 2)
    plt.title("Inverse color")
    plt.imshow(inversed_image)
    plt.show()


username = input("Enter your username: ")
if username in pic_map:
    img = pic_map[username]
    print(f"Welcome {username}, you can now perform operations on your image")
    print("1. Addition of images")
    print("2. Adjust Brightness & Contrast")
    print("3. Sharpening")
    print("4. Removing Noise")
    print("5. Image Scaling")
    print("6. Inverse Transformation")
    operation = int(input("Enter the number of the operation you want to perform: "))

    if operation == 1:
        addition_of_images(img)
    elif operation == 2:
        adjust_brightness_contrast(img)
    elif operation == 3:
        sharpening(img)
    elif operation == 4:
        removing_noise(img)
    elif operation == 5:
        image_scaling(img)
    elif operation == 6:
        inverse_transformation(img)
```
